### 뉴스 제목 추출하기

```python
import requests
from bs4 import BeautifulSoup
```

```python
req_params = {
    'sid': 100 # 정치
}

r = requests.get('https://news.naver.com/section/{sid}'.format(**req_params))

# sa_text_title > strong
soup = BeautifulSoup(r.text, 'html.parser')
sa_titles = soup.select('.sa_text_title > strong')

sa_titles_text = [title.get_text() for title in sa_titles]
print('\n'.join(sa_titles_text))
```
